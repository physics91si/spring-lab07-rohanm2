# Lecture 07 - SciKit Image

SciKit Image is a collection of algorithms for image processing!

```python
import time
```

```python
start_time = time.clock()

from skimage import data, io, filters
from matplotlib import pyplot as plt
import numpy as np

print(time.clock() - start_time, "seconds")
```

```python
cat = data.chelsea()
print(type(cat))
print(cat.shape) #This shows that cat is a 300-by-451 pixel image with three channels (red, green, and blue). 
print('Min:',cat.min())
print('Max:',cat.max())
io.imshow(cat)
plt.show()
```

```python
cup = data.coffee()
print('Min:',cup.min())
print('Max:',cup.max())
io.imshow(cup)
plt.show()
```

```python
image = data.coins()
io.imshow(image)
plt.show()
```

```python
edges = filters.sobel(image)
io.imshow(edges)
plt.show()
```

```python
moon = data.moon()
io.imshow(moon)
plt.show()

nrows, ncols = moon.shape
row, col = np.ogrid[:nrows, :ncols]
cnt_row, cnt_col = nrows / 2, ncols / 2
outer_disk_mask = ((row - cnt_row)**2 + (col - cnt_col)**2 > (nrows / 2)**2)
moon[outer_disk_mask] = 0
io.imshow(moon)
plt.show()
```

```python
from skimage.color import rgb2gray

original = data.astronaut()
io.imshow(original)
plt.show()

grayscale = rgb2gray(original)
io.imshow(grayscale)
plt.show()
```

```python
from skimage.morphology import skeletonize
from skimage.util import invert
```

```python
image = data.horse()
io.imshow(image)
plt.show()
```

```python
# Invert the horse image as skeleton is made of whitespace
image = invert(data.horse())
io.imshow(image)
plt.show()
```

```python
# perform skeletonization
skeleton = skeletonize(image)
io.imshow(skeleton, cmap=plt.cm.gray)
plt.show()
```
